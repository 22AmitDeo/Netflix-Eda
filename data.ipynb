```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("netflix_titles.csv")
```

```python
df['listed_in']=df['listed_in'].str.split(', ')
```

```python
all_genres = df.explode('listed_in')  
genre_counts = all_genres['listed_in'].value_counts()
print(genre_counts.head(10))
```

```python
top_directors=df['director'].dropna().value_counts()
print(top_directors.head(10))
```

```python
df['release_year'] = df['release_year'].astype(int)
content_trend = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
plt.plot(content_trend.index, content_trend.values, marker='o', linestyle='-')
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.title("Netflix Content Trend Over Time")
plt.grid()
plt.show()

```
